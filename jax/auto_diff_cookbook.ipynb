```python
import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random
import matplotlib.pyplot as plt
from typing import Callable
key = random.key(0)
```

```python
def plus_1(x): 
    return x+1
def basic_quadratic(x, a=1, b=0, c=0): 
    return a*x**2 + b*x + c

def basic_cubic(x, a=1, b=0, c=0, d=0):  
    return a*x**3 + b*x**2 + c*x + d

def reciprocal(x):
    return 1/x

def reciprocal_sq(x): 
    return 1/(x**2)
def sigmoid(x):
    return 0.5 * (jnp.tanh(x / 2) + 1)


    
```

```python
def plot_grad(x:jnp.array, func:Callable): 
    func_grad = grad(func)
    y = []
    y_grad = []

    for val in x: 
        y.append(func(val))
        y_grad.append(func_grad(val))
    fig, plts = plt.subplots(nrows=1, ncols=2)
    plts[0].plot(x, y)
    plts[0].set_title(f" {func.__name__}") 
    plts[1].plot(x, y_grad)
    plts[1].set_title(f"Grad of {func.__name__}") 
    fig.tight_layout()
    plt.show()
       
```

```python
x = jnp.linspace(-5,5,2001)
plot_grad(x, sigmoid)
```
