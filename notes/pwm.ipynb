# PWM Signals

Take a periodic pulse wave f(t) with period T, y_min and y_max values, with a constant duty cycle

The average would be the integral of this period 0->T , divided by 1/T 



As $f(t)$ is a pulse wave, its value is $y_{\text{max}}$ for  
$0 < t < D \cdot T$ and $y_{\text{min}}$ for $D \cdot T < t < T$.  
The above expression then becomes:

$$
\bar{y} = \frac{1}{T} \left( \int_0^{DT} y_{\text{max}} \, dt + \int_{DT}^{T} y_{\text{min}} \, dt \right)
$$

$$
= \frac{1}{T} \left( D \cdot T \cdot y_{\text{max}} + T(1 - D) \cdot y_{\text{min}} \right)
$$

$$
= D \cdot y_{\text{max}} + (1 - D) \cdot y_{\text{min}}
$$
